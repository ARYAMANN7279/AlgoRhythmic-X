```python
def log_std(obj):
  # list_ans=[]
  # for i in range(len(obj)):
  #   if(i!=0):
  #     b=np.log((obj['Close'].iloc[i]/obj['Close'].iloc[i-1]))
  #     list_ans.append(b)
  # fin=np.array(b)
  # x=fin.std()
  # return x*np.sqrt(252)

  a=np.log(obj['Close'])-np.log(obj['Close'].shift(-1))
  std1=a.std()
  return std1*np.sqrt(len(obj))
```

```python
def return_m(obj_m,type1):
  ret=(obj_m['Close'].iloc[-1]/obj_m['Close'].iloc[type1])-1
  return ret
```

```python
def momentum_ratio(obj):
  std=log_std(obj[1])
  ret_m_12=return_m(obj[0],-13)
  ret_m_6=return_m(obj[0],-7)
  ratio_6=ret_m_6/std
  ratio_12=ret_m_12/std
  return ratio_6,ratio_12
```

```python
def mean_std_all_ratio_6and12(obj_all):
  list1=[]
  list2=[]
  for i in range(len(obj_all)):
    r_6,r_12=momentum_ratio(obj_all[i])
    list1.append(r_6)
    list2.append(r_12)
  a1=np.array(list1)
  a2=np.array(list2)
  return a1.mean(),a1.std(),a2.mean(),a2.std()
```

```python
for i in range(len(stock_data)):
  print(str(i)+ " "+ str(momentum_ratio(stock_data[i])))
```

```python
mean_6,std_6,mean_12,std_12=mean_std_all_ratio_6and12(stock_data)
```

```python
def weightedZscore(obj):
  r_6,r_12=momentum_ratio(obj)
  Z_12=(r_12-mean_12)/std_12
  Z_6=(r_6-mean_6)/std_6
  return (Z_6+Z_12)/2
```

```python
def Zscore_w_n(obj):
  a=weightedZscore(obj)
  if(a>=0):
    return (1+a)
  else:
    return (1-a)**(-1)
```

```python
#MAIN function
dict_f={}
for i in range(len(stock_data)):
  ZScore=Zscore_w_n(stock_data[i])
  name=stocks[i]
  dict_f.update({name:ZScore})


ans=dict(sorted(dict_f.items(), key=lambda item: item[1]))
iter=0
list_ans=[]
for key,value in ans.items():
  if(iter>118):
    print(key +" : "+str(value))
    list_ans.append(key)
  iter+=1


```
